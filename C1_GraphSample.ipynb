# **시각화 예제**

```python
import numpy as np
import matplotlib.pyplot as plt

# 0부터 5까지 0.2씩 증가하는 값을 저장
t = np.arange(0, 5, 0.2)
t2 = t ** 2
t3 = t ** 3

# avg를 평균으로, sigma를 표준편차로 갖는 정규분포로부터 랜덤하게 값 추출
avg, sigma = 100, 15
x = avg + sigma * np.random.randn(10000)
```

# **선형 그래프**

```python
plt.plot(t, t2, label="t2")
plt.plot(t, t3, label="t3")
plt.title("graph")
plt.legend()
plt.show()
```

# **서브 플롯**

```python
# (row, column, index)
plt.subplot(1, 2, 1)
plt.plot(t, t)
plt.subplot(1, 2, 2)
plt.plot(t, t3)
plt.show()
```

# **히스토그램**

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('histogram')
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.show()
```
